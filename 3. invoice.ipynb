# To draw rectangle for date

check the format

here mm/dd/yyyy

To check the pattern we need to import regular expression

```python
import re
```

7/19/2018

mm/dd/yyyy

regular expression for date:

\d{1,2}/\d{1,2}/\d{4} => 1,2 means the date and month should be read when it is 1 digit and 2 digit and year read when 4 digit. d is used to collect the datas such as month, day, year.

```python
length = len(dict['text'])
length
```

```python
for i in range(length):
    if re.search(r'(\d{1,2}/\d{1,2}/\d{4})', dict['text'][i]):  # search, match can be used. r is used to read the rehilar expression
        x=dict['left'][i]
        y=dict['top'][i]
        w=dict['width'][i]
        h=dict['height'][i]
        cv2.rectangle(invoice_img,(x,y),(x+w,y+h),(0,0,255),2)
        print(dict['text'][i])
cv2.imshow("Invoice",invoice_img)
cv2.waitKey()
cv2.destroyAllWindows()
```
